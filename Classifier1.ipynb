## Threshold-based Classifier

```python
import matplotlib.pyplot as plt

C1 = [(2,3), (3,3), (3,4), (1,4), (4,1), (4,3)]
C2 = [(0,0), (0,3), (1,1), (1,2), (2,1), (2,2)]

def plot(thx, thy):  #Part 1
  x = [point[0] for point in C1]
  y = [point[1] for point in C1]
  x2 = [point[0] for point in C2]
  y2 = [point[1] for point in C2]
  plt.figure(figsize=(8, 6))
  plt.scatter(x, y, color="red", marker=">", label="C1")
  plt.scatter(x2, y2, color="blue", marker=(5, 1), label="C2")
  plt.axvline(x=thx, color="black", linestyle="--", label="thx")
  plt.axhline(y=thy, color="black", linestyle="--", label="thy")
  plt.title("Threshold-based Classifier")
  plt.xlabel('X - feature 1')
  plt.ylabel('Y - feature 2')
  plt.legend()
  plt.xlim(-1, 5)
  plt.ylim(-1, 5)
  plt.show()

def threshvals(prompt):  #Part 2
  while True:
    threshold_input = input(prompt)
    if threshold_input.lower() == 'x':
      return 'x'  #Part 7
    try:
      threshold = float(threshold_input) #Part 3
      return threshold
    except ValueError:
      print("Error: Please type in an integer or 'x'.")

def classification(thx, thy, point):
  x, y = point
  if x >= thx and y >= thy:
    return "C1"
  else:
    return "C2"

def accuracy(C1, C2, thx, thy):  #Part 4
  correct_classifications = 0
  total_samps = len(C1) + len(C2)
  for point in C1:
    if classification(thx, thy, point) == "C1":
      correct_classifications += 1
  for point in C2:
    if classification(thx, thy, point) == "C2":
      correct_classifications += 1

  accuracy_val = correct_classifications / total_samps
  return accuracy_val

def main():  #Part 5 and 6
  while True:
    thx = threshvals("Enter x threshold value: ")
    if thx == "x":  #Part 7
      break
    thy = threshvals("Enter y threshold value: ")
    if thy == "x":  #Part 7
      break
    accuracy_two = accuracy(C1, C2, thx, thy)
    print(f"Training accuracy with the thresholds of {thx}, {thy}: {accuracy_two*100:.2f}%")
    plot(thx, thy)

if __name__ == "__main__":
  main()
```

#### **Part 1**:  To plot the data samples I created a function that would take the parameters thx and thy that would assign x and y to C1 and x2 and y2 to C2 by iterating through each point[0] and point[1] with for loops.  Next x and y were marked with > to denote a triangle shape and given the color red.  Following x2 and y2 were marked with (5, 1) to give them a star shape and rendered the color blue.

#### **Part 2 and 3**: To obtain the user input threshold values I created a function that took a prompt as its argument.  We store the functions statements within an infinite loop that only breaks if 'x' is input by the user in the main.  The function serves to convert the user input into a float and if it is successful it returns the float value.  If the input is anything else besides an integer or 'x' the ValueError gets raised and prompts the user to input again.

#### **Part 4**:  To calculate the accuracy I created the accuracy function taking C1, C2, thx, and thy as its parameters.  The function begins by initializing a counter to equal 0 to start.  Following we get the total number of samples from the sample data by using the len() function on both C1 and C2 arrays and adding them together.  Next a for loop is used to loop through each point in C1 and C2 and uses the classification function to determine the predicted class of each point.  If the classification is correct then the initial counter goes up by 1.  After the loop finishes the variable accuracy_val divides the correct classifications by the total number of samples to get and return the training accuracy.

#### **Part 5, 6, and 7**: plt.axvline() and plt.axhline() were added to the plot function to mark the threshold lines on the graph.  That is why I had thx and thy as paremeters in the plot function.  Next the main function allows for all of the functions to run in tandem with an infinite loop allowing the user to enter different sets of thresholds over and over unless 'x' is input by the user.  The function also serves to define thx and thy.

#### **Part 8**: Based on my observations the best set of thresholds that gives the highest accuracy is (2,3) or (0,3) which both have a training accuracy of 83.33%.  Other thresholds close were (3,3), (3,2), (2,2), and (1,2) which had a training accuracy of 75.00% exact.
